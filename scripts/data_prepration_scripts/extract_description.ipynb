```python
import pandas as pd
import pickle
```

```python
data_path = "../../../data/"
jobs = pd.read_csv(data_path + "software_engineering_description.csv.csv")
jobs
```

```python
jobs = jobs.fillna("")
jobs.loc[:,"responsibilities"] = map(lambda x: "" if x=="[]" else x, jobs["responsibilities"])
jobs.loc[:,"skills"] = map(lambda x: "" if x=="[]" else x, jobs["skills"])
```

```python
jobs_text_list  = list( jobs["title"].str.cat([jobs["normalizedTitle"],jobs["responsibilities"],jobs["skills"],jobs["jobDescription"]], sep = " "))
```

```python
pickle.dump(jobs_text_list, open(data_path + "jobs_raw.pkl", "wb"))
```
